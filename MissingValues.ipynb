```python
import numpy as np
import pandas as pd
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()  # isna means jo null value hai vahan true aa jaega baki place par false
```

```python
df.isna().sum()  #.sum lagane se false(null value) ki counting aa jaegi in column
```

```python
df.isna().any()   #.any lagane se ye pata lagta ki koi aisa column hai jisme null value na ho. vahan False likha aa jaega
```

**Removing Missing Data**

```python
df
```

```python
df.dropna() # null value wale columns delete ho jaeyenge
```

```python
df
```

```python
df.dropna(thresh=3)   # thresh means ki hm threshold laga re for instance 3 ki atleast 3 non Nan values honi chahiye in rows
```

**Filling the missing Data**

```python
df
```

```python
df.fillna(0)   # Nan value ko 0 se fill krna
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())  # if we want to fill the Nan values with the mean value
```
