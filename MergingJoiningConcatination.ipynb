```python
import numpy as np
import pandas as pd
```

**Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries) # commom ones will merge
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner')
# on means jiske base par ap merge krna chahte vo on mai k=likh skte like employee_id ke base par
# how = inner means jo apne on wala base liya hai uske andar common things ko merge kr dena
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')
# yahan how = outer means employee_id ko pura consider krna na ki sirf common walo ko
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')
# left means employees ke base par
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
# right means salaries ke base par
```

**Concatination of 2 dataframes**

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # yahan automatically axis =0 yani column ke base par cancat hoga
```

```python
pd.concat([df1,df2],axis= 1) # axis = 1 to row ke basis par concat hoga
```

**Joining 2 dat**a **frames**

```python
df1 = pd.DataFrame({
    'name': ['Alice','Bob','Charlie']
},index =[1,2,3])

#second data frame

df2 = pd.DataFrame({
    'score': [85,90,75]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```
